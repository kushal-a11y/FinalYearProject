```python
df["UserID"] = [i for i in range(1, len(df) + 1)]
df.set_index("UserID", inplace=True)

# Generate castes once
castes_generated = [select_caste_by_weight() for _ in range(len(df))]

user_profile = {
    "Age": df["Age_at_Marriage"],
    "Gender": df["Gender"],
    "Education": [select_degree_by_weight() for _ in range(len(df))],
    "Caste": castes_generated,
    "Profession": [slect_profession_by_weight() for _ in range(len(df))],
    "Religion": [
        "Hindu" if caste in hindu_castes else
        random.choices(
            population=list(religion_weights.keys()),
            weights=list(religion_weights.values())
        )[0]
        for caste in castes_generated
    ],
    "Residence": [random.choice(['urban', 'rural']) for _ in range(len(df))],
    "Height_cm": [random.choice(heights) for _ in range(len(df))],
    "Extras": df.apply(
        lambda x: json.dumps({
            "Urban_Rural": x["Urban_Rural"],
            "Spouse_Working": x["Spouse_Working"]
        }),
        axis=1
    )
}

user_profile_df = pd.DataFrame(user_profile)
user_profile_df.head(5)
```

```python
def assign_preferred_cast(row):
    if row["Caste_Match"] == "Same":
        return user_profile_df.loc[row.name, "Caste"]
    else:
        return select_caste_by_weight()
```

```python
def assign_preferred_religion(row):
    if row["Inter-Religion"] == "No":
        print(row.name)
        return user_profile_df.loc[row.name, "Religion"]
    else:
        return random.choices(
            population=list(religion_weights.keys()),
            weights=list(religion_weights.values())
        )[0]
```

```python
sum = df["Age_at_Marriage"].value_counts().sum()
age_weights = dict(df["Age_at_Marriage"].value_counts().apply(lambda x : x/sum))

print(age_weights)
```

```python
gender = df["Gender"].apply(lambda x: "Female" if x == "Male" else "Male")
```

```python
user_preference_df = pd.DataFrame(index=df.index)
user_preference_df["PreferenceID"] = [f"P{i:03d}" for i in range(1,len(user_profile_df) + 1)]
user_preference_df.set_index("PreferenceID", inplace=True)
user_preference_df["Age"] = [random.choices(population=list(age_weights.keys()),
                                             weights=list(age_weights.values()))[0] 
                             for _ in range(len(df))]
user_preference_df["Gender"] = gender.values
user_preference_df["Education"] = [select_degree_by_weight() for _ in range(len(df))]
user_preference_df["Caste"] = [
    user_profile_df.loc[i, "Caste"] if df.loc[i, "Caste_Match"]=="Same" else select_caste_by_weight()
    for i in df.index
]
user_preference_df["Profession"] = [slect_profession_by_weight() for _ in range(len(df))]
user_preference_df["Residence"] = [random.choice(['urban','rural']) for _ in range(len(df))]
user_preference_df["Religion"] = [
    user_profile_df.loc[i, "Religion"] if df.loc[i, "Inter-Religion"]=="No" else
    random.choices(population=list(religion_weights.keys()), weights=list(religion_weights.values()))[0]
    for i in df.index
]
user_preference_df["Height_cm"] = [random.choice(heights) for _ in range(len(df))]
user_preference_df["Extras"] = df.apply(
    lambda x: json.dumps({
        "Urban_Rural": x["Urban_Rural"],
        "Spouse_Working": x["Spouse_Working"]
    }),
    axis=1
)
user_preference_df["userID"] = user_profile_df.index

user_preference_df.head(5)
```
